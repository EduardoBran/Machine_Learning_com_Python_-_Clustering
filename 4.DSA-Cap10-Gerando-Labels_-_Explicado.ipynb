# üìÑ Descri√ß√£o Geral
O script utiliza o **K-Means** para segmentar dados do conjunto **Iris** em clusters. A partir do modelo de clustering, o c√≥digo aplica o **PCA** para reduzir a dimensionalidade e possibilitar a visualiza√ß√£o dos clusters em 2D. O script ainda gera **labels** para os clusters e permite identificar quais dados pertencem a qual cluster.

---

# üéØ Objetivo
O objetivo do script √©:

1. **Segmentar** dados do conjunto **Iris** em 3 clusters utilizando o algoritmo **K-Means**.
2. **Visualizar** os clusters em 2D ap√≥s a aplica√ß√£o do **PCA**.
3. **Identificar** a quais clusters pertencem os dados e atribuir **labels** a eles.

<br><br>

---

<br>

# üî¢ Bloco 1: Importa√ß√£o das Bibliotecas e Carregamento do Dataset üìö

```python
# Imports
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pylab as pl
```

### üîç Objetivo:
- Importar as **bibliotecas** necess√°rias para o **clustering**, **PCA** e **visualiza√ß√£o**.
- Preparar o ambiente para carregar o conjunto de dados **Iris**.

---

# üìä Bloco 2: Carregando e Visualizando os Dados üîé

```python
# Carregando o dataset
iris = load_iris()

# Visualizando todo o conte√∫do
iris
```

### üîç Objetivo:
- Carregar o conjunto de dados **Iris** (sem r√≥tulos, apenas as caracter√≠sticas de entrada).
- Visualizar a estrutura dos dados carregados para entender suas dimens√µes e atributos.

---

# üîç Bloco 3: Criando o Modelo K-Means e Ajustando os Dados ‚öôÔ∏è

```python
# Criando o modelo
kmeans = KMeans(n_clusters = 3, random_state = 111)
kmeans.fit(iris.data)
```

### üîç Objetivo:
- Criar e treinar o modelo **K-Means** com 3 clusters.
- Ajustar o modelo aos dados do conjunto **Iris** para identificar os clusters.

---

# üó∫Ô∏è Bloco 4: Criando o Cluster Map üåç

```python
# Criando o Cluster Map
cluster_map = pd.DataFrame(iris.data)
cluster_map['cluster'] = kmeans.labels_
cluster_map
```

### üîç Objetivo:
- Criar um **Cluster Map** que adicione a coluna de **labels** (`cluster`) gerados pelo modelo **K-Means** ao conjunto de dados **Iris**.
- Visualizar como os dados s√£o distribu√≠dos entre os clusters identificados.

---

# üè∑Ô∏è Bloco 5: Filtrando os Dados Pelo Cluster üîç

```python
# Filtrando os Dados Pelo Cluster
cluster_map[cluster_map.cluster == 2]
```

### üîç Objetivo:
- Filtrar e visualizar os dados pertencentes ao **Cluster 2**.
- Verificar como os dados est√£o organizados dentro dos clusters.

---

# üîÑ Bloco 6: Reduzindo a Dimensionalidade com PCA üîÑ

```python
# Reduzindo a dimensionalidade
pca = PCA(n_components = 2).fit(iris.data)
pca_2d = pca.transform(iris.data)
```

### üîç Objetivo:
- Aplicar o **PCA** para reduzir os dados de 4 dimens√µes para 2, facilitando a visualiza√ß√£o gr√°fica dos clusters em um plano bidimensional.

---

# üé® Bloco 7: Gerando Labels e Visualizando os Clusters em 2D

```python
# Gerando "labels" para os resultados dos clusters
for i in range(0, pca_2d.shape[0]):
    if kmeans.labels_[i] == 0:
        c1 = pl.scatter(pca_2d[i,0],pca_2d[i,1], c ='r', marker = '+')
    elif kmeans.labels_[i] == 1:
        c2 = pl.scatter(pca_2d[i,0],pca_2d[i,1], c ='g', marker = 'o')
    elif kmeans.labels_[i] == 2:
        c3 = pl.scatter(pca_2d[i,0],pca_2d[i,1], c ='b', marker = '*')
        pl.legend([c1, c2, c3],['Cluster 0', 'Cluster 1', 'Cluster 2'])
        pl.title('Clusters K-means com Iris dataset em 3 clusters')
pl.show()
```

### üîç Objetivo:
- Gerar **labels** visuais para cada cluster e atribuir **cores** e **marcadores** espec√≠ficos para cada grupo.
- Visualizar os clusters no gr√°fico 2D ap√≥s a redu√ß√£o da dimensionalidade com **PCA**.

---

# üìù Bloco 8: Visualizando os Resultados em Texto üìú

```python
# Exibindo os resultados dos clusters de forma textual
cluster_results = pd.DataFrame(iris.data, columns=iris.feature_names)
cluster_results['Cluster'] = kmeans.labels_

# Exibindo as primeiras 10 linhas para ver como os dados est√£o segmentados por cluster
cluster_results.head(10)
```

### üîç Objetivo:
- Criar um **DataFrame** que contenha as caracter√≠sticas dos dados, bem como os **labels** dos clusters atribu√≠dos pelo modelo **K-Means**.
- Exibir as primeiras 10 linhas para visualizar a segmenta√ß√£o dos dados de forma textual, observando a qual cluster cada amostra pertence.
